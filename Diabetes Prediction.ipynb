```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
```

```python
x=data.drop(columns='Outcome',axis=1)
```

```python
y=data['Outcome']
```

```python
x
```

```python
x.shape
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
standard_data=scaler.transform(x)   #scaler.fit_transform()
```

```python
print(standard_data)
```

```python
x=standard_data
y=data['Outcome']
```

```python
print(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation

```python
x_train_pred = model.predict(x_train)
```

```python
training_acc = accuracy_score(x_train_pred,y_train)
```

```python
print("Accuracy score on training data: ",training_acc)
```

```python
x_test_pred = model.predict(x_test)
```

```python
testing_acc = accuracy_score(x_test_pred,y_test)
```

```python
print("Accuracy score on testing data: ",testing_acc) 
```
