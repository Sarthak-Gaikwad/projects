# Exploratory Data Analysis on Airbnb Listings

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter settings
%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('/Users/saurabh/Downloads/Sarthak/Placement/Projects/listings.csv')
df.head()
```

```python
print("Shape:", df.shape) #shape of the dataset
```

```python
print("\nColumns:", df.columns.tolist()) # Column names
```

```python
df.describe()
```

```python
df.isnull().sum()  #check missing values
```

```python
df = df[['id', 'name', 'host_id', 'neighbourhood', 'room_type',
         'price', 'minimum_nights', 'number_of_reviews', 
         'reviews_per_month', 'availability_365']]
```

```python
df.dropna(subset=['price', 'reviews_per_month'], inplace=True)
```

```python
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float) # Convert price to numeric (remove $ sign, commas)
```

```python
df = df[df['price'] < 500]  # Removing extreme price listings
```

# Univariate Analysis

```python
# Histogram of prices
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True,color='blue')
plt.title('Distribution of Airbnb Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Room type counts
plt.figure(figsize=(6,4))
sns.countplot(x='room_type', data=df,color='blue')
plt.title('Room Type Distribution')
plt.show()
```

# Bivariate Analysis


```python
# Price vs Room Type
plt.figure(figsize=(6,4))
sns.boxplot(x='room_type', y='price', data=df,color='black')
plt.title('Price Variation by Room Type')
plt.show()
```

```python
# Price vs Availability
plt.figure(figsize=(6,4))
sns.scatterplot(x='availability_365', y='price', alpha=0.5, data=df,color='red')
plt.title('Availability vs Price')
plt.show()
```

# Multivariate Analysis

```python
# Grouping by Neighbourhood
neigh_price = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)

# Plot Top 10 Expensive Neighbourhoods
plt.figure(figsize=(10,5))
neigh_price.head(10).plot(kind='bar',color='blue')
plt.title('Top 10 Neighbourhoods by Average Price')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.show()
```

# Correlation Heatmap

```python
# Only select numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

correlation_matrix = numeric_df.corr()

# Plotting the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
